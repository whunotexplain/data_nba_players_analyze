$c^2 = a^2 + b^2$

где:
- **c** — гипотенуза
- **a**, **b** — катеты

Гипотенуза вычисляется как квадратный корень из суммы квадратов катетов:

$$ c = \sqrt{a^2 + b^2} $$

```python
import math

katet_1 = 15
katet_2 = 20

gipotenuza = math.sqrt(pow(katet_1, 2) + pow(katet_2, 2))

gipotenuza

```
