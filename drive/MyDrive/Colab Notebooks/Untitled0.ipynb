Создайте базовый класс Employee (Сотрудник) со следующими атрибутами и методами:

Атрибуты:

name - имя сотрудника
salary - зарплата
Методы:

work() - выводит на экран сообщение: "Работаю"
take_break() - выводит на экран сообщение: "Отдыхаю"

```python
class Employee:
  def __init__(self, name, salary) -> None:
    self.name = name
    self.salary = salary

  def work(self):
    print("Работаю")

  def take_break(self):
    print("Отдыхаю")

class Developer(Employee):
  def __init__(self, name, salary, programming_language) -> None:
     self.programming_language = programming_language
     super().__init__(name, salary)

  def write_code(self):
    print("Пишу код")

if __name__ == "__main__":
    empl = Employee("Egor", 6000)
    dev = Developer("Egor", 6000, "Python")
    empl.work()
    empl.take_break()
    dev.write_code()


```

```python
alice_hobbies = {"рисование", "программирование", "чтение", "йога"}
marusya_hobbies = {"фотография", "программирование", "йога", "путешествия"}

print(f"Общие увлечения: {alice_hobbies & marusya_hobbies}")
print(f"Только у Алисы: {alice_hobbies - marusya_hobbies}")
print(f"Все увлечения: {alice_hobbies | marusya_hobbies}")

hobby = "плавание"
if hobby in alice_hobbies | marusya_hobbies:
    print(f"'{hobby}' есть у кого-то из друзей")
else:
    print(f"'{hobby}' нет ни у кого")
```

```python
prices = {"хлеб": 50, "молоко": 80, "яйца": 120, "сыр": 200, "колбаса": 300, "яблоки": 70, "бананы": 120}

print(f"Общая стоимость: {sum(prices.values())} руб.")

most_expensive = max(prices, key=prices.get)
print(f"Самый дорогой: {most_expensive} - {prices[most_expensive]} руб.")


cheap = [p for p, c in prices.items() if c < 100]
print(f"Продукты дешевле 100 руб.: {', '.join(cheap)}")
```

```python
point_a = (2, 5)
point_b = (8, 1)

length = abs(point_b[0] - point_a[0])
width = abs(point_b[1] - point_a[1])

print(f"Длина: {length}, Ширина: {width}")
print(f"Периметр: {2 * (length + width)}")
print(f"Площадь: {length * width}")
```

```python
marks = input().split()

marks_from_map = list(map(int, marks))
marks_from_comprehension = [int(x) for x in marks]

marks_list = marks_from_comprehension

avg = sum(marks_list) / len(marks_list)

for m in marks_list:
    if m > avg:
        print(m)
```
