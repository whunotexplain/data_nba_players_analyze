##Binary Search (Бианрный поиск)

```python
def my_binary_search(nums: list, num: int) -> int:
  low = 0
  high = len(nums) - 1

  while low <= high:
    mid = (low + high) // 2
    if nums[mid] == num:
      return mid
    if nums[mid] > num:
      high = mid - 1
    else:
      low = mid + 1
  return None

if __name__ == "__main__":
  my_list = [1,2,3,4,5,6,7,8,9]
  print(my_binary_search(my_list, 3))
  print(my_binary_search(my_list, 13))
```

Задание LeetCode, по посчету отрицательных чисел в трехмерной матрице размером m x n

```python


class Solution:
    def countNegatives(self, grid: List[List[int]]) -> int:
        count = 0
        m = len(grid)
        n = len(grid[0])

        for row in range(m):
            left = 0
            right = n - 1
            first_neg = n

            while left <= right:
                mid = (left + right) // 2
                if grid[row][mid] < 0:
                    first_neg = mid
                    right = mid - 1
                else:
                    left = mid + 1

            count += (n - first_neg)
        return count
```
